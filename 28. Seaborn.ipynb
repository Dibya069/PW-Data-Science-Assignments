## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter Plot:
 - Use: Scatter plots are used to visualize the relationship between two continuous variables. They show how one variable changes with respect to another variable.
2. Line Plot:
 - Use: Line plots are used to display the trend or pattern of a continuous variable over a continuous axis, such as time. They are useful for visualizing time series data or any data with a natural ordering.
3. Bar Plot:
 - Use: Bar plots are used to compare and display the values of categorical variables. They can show the distribution of a categorical variable or compare the values of a numerical variable across different categories.
4. Histogram:
 - Use: Histograms are used to visualize the distribution of a single variable. They provide insights into the shape, spread, and central tendency of the data.
5. Heatmap:
 - Use: Heatmaps are used to display tabular data where the values are represented by colors. They are commonly used to visualize correlation matrices or to highlight patterns or relationships in data.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri_data = sns.load_dataset("fmri")
```

```python
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and axis labels
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
titanic_data = sns.load_dataset("titanic")
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()

plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack')
plt.title('Histogram of Price by Cut')
plt.xlabel('Price')
plt.ylabel('Count')

# Show the plot
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris_data = sns.load_dataset("iris")

sns.pairplot(data=iris_data, hue='species')
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_data = sns.load_dataset("flights")
flights_matrix = flights_data.pivot("month", "year", "passengers")

sns.heatmap(data=flights_matrix, cmap="YlGnBu")
plt.show()
```
