```python
def remove_punc(text):
    return text.translate(str.maketrans('', '', exclude))
```

## Chart Conversioin Handeling

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible',
    'WRT':'With Respect To',
    'IDK': 'I Do not know'
}
```

```python
def chat_conversion(text):
    new_text = []
    count = []
    c = 1

    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
            c += 1
            count.append(c)
        else:
            new_text.append(w)

        texxt = " ".join(new_text)
    return texxt, count
```

```python
twe2 = []
oc = []

for i in range(0, len(df)):
  t, c = chat_conversion(df['tweet'][i])
  twe2.append(t)
  oc.append(c)

df["tweet"] = pd.DataFrame(twe2)

print(oc)
print(df.head())
```

## incorrect_text handling

```python
c = 0

for i in range(0, len(df)):
  a = df["tweet"][i]

  cor1 = TextBlob(a).correct()
  cor2 = str(cor1)
  df.at[i, "tweet"] = cor2

  if a != cor2:
    c += 1

print(c)
print(df.head())
```

## Stop Words

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
for i in range(0, len(df)):
  df["tweet"][i] = remove_stopwords(df["tweet"][i])

df.head()
```
